```
import pandas
import matplotlib.pyplot as plt
```

```
# set up variables, incl x y as lists for successive values
x = [0.6]
y = [0.7]

lr = 0.1
```

```
# run gradient descent for a while

for i in range(199):
  dx = lr * y[i]
  dy = - lr * x[i]
  x.append( x[i] + dx )
  y.append( y[i] + dy )
  pass
```

```
# plot successive values of x and y

df = pandas.DataFrame(data=list(zip(x,y)), columns=['x','y'])

df.plot(figsize=(10,6))
```

```
df.plot.scatter('x','y', figsize=(10,10))
```

```
fig, axes = plt.subplots(nrows=1, ncols=2)
df.plot(figsize=(20,6),ax=axes[0])
axes[0].set_xlabel("iteration")
df.plot.scatter('x','y', figsize=(20,6), ax=axes[1])
```
